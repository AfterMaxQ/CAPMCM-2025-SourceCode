```python
from graphviz import Digraph
import os

# 配置路径
OUTPUT_DIR = r"G:\jupyter\2025APMCM\Cleaned_Data"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 创建有向图
dot = Digraph(comment='Modeling Framework', format='png')
# 增加画布尺寸以容纳更大的文字
dot.attr(rankdir='TB', size='12,15', dpi='300', compound='true', splines='ortho')

# 全局节点样式 - 字体再次放大
dot.attr('node', shape='box', style='filled', fontname='Helvetica', fontsize='26', margin='0.3,0.2')
dot.attr('edge', fontname='Helvetica', fontsize='18', color='#555555')

# --- 1. Data Layer (灰色背景) ---
with dot.subgraph(name='cluster_0') as c:
    # 标题字体放大
    c.attr(label='Phase I: Data Forensics & Shock Construction (Sec 3)', style='filled', color='#f2f2f2', fontsize='32', fontname='Helvetica-Bold')
    c.node('Raw', 'Raw Data Sources\n(USITC, FRED, UN Comtrade)', fillcolor='#e1e1e1')
    c.node('Clean', 'ETL & Anomaly Detection\n(Isolation Forest)', fillcolor='#ffffff', color='#333333')
    c.node('Baseline', 'Zero-Tariff Baseline Verification\n(Mean Rate ≈ 3.78%)', fillcolor='#ffffff', color='#333333')
    c.node('Shock', 'Policy Shock Function S(·)\n(Constructed 10% Baseline)', shape='hexagon', fillcolor='#ffddc1', color='#d35400')
    
    c.edge('Raw', 'Clean')
    c.edge('Clean', 'Baseline')
    c.edge('Baseline', 'Shock', label='Regime Shift')

# --- 2. Sectoral Modeling Layer (分组) ---
with dot.subgraph(name='cluster_1') as c:
    # 标题字体放大
    c.attr(label='Phase II: Sectoral Impact Analysis (Micro-Level)', style='dashed', color='#2980b9', fontsize='32', fontname='Helvetica-Bold')
    
    # Q1 Soybean
    c.node('Q1_Model', 'Q1: Soybean Dynamics\n(Lotka-Volterra Model)', fillcolor='#d4e6f1')
    c.node('Q1_Out', 'Trade Diversion & Stagnation\n(Phase Space Trajectory)', fillcolor='#ffffff', style='dashed')
    
    # Q2 Auto
    c.node('Q2_Model', 'Q2: Auto Supply Chain\n(VARX & Cost Matrix)', fillcolor='#d4e6f1')
    c.node('Q2_Out', 'Price Transmission Lag\n& Tariff-Jumping FDI', fillcolor='#ffffff', style='dashed')
    
    # Q3 Chips
    c.node('Q3_Model', 'Q3: Chip Sovereignty\n(Game Theory & Dual-Axis)', fillcolor='#d4e6f1')
    c.node('Q3_Out', 'Security Paradox\n& Price-Capacity Divergence', fillcolor='#ffffff', style='dashed')

    # Edges internal
    c.edge('Q1_Model', 'Q1_Out')
    c.edge('Q2_Model', 'Q2_Out')
    c.edge('Q3_Model', 'Q3_Out')

# --- 3. Macro Modeling Layer ---
with dot.subgraph(name='cluster_2') as c:
    # 标题字体放大
    c.attr(label='Phase III: Macroeconomic Aggregation (Macro-Level)', style='dashed', color='#27ae60', fontsize='32', fontname='Helvetica-Bold')
    
    # Q4 Fiscal
    c.node('Q4_Model', 'Q4: Fiscal Revenue\n(Hybrid SARIMAX-LSTM)', fillcolor='#d5f5e3')
    c.node('Q4_Out', 'Dynamic Laffer Curve\', fillcolor='#ffffff', style='dashed')
    
    # Q5 Macro
    c.node('Q5_Model', 'Q5: Reshoring & Economy\n(Entropy Index & Macro VAR)', fillcolor='#d5f5e3')
    c.node('Q5_Out', 'Stagflation Risk\n& "False Reshoring"', fillcolor='#ffffff', style='dashed')

    c.edge('Q4_Model', 'Q4_Out')
    c.edge('Q5_Model', 'Q5_Out')

# --- 4. Conclusion ---
# 结论节点字体放大
dot.node('Final', 'Policy Evaluation Conclusion\n(Suboptimal Outcome)', shape='Mdiamond', fillcolor='#fad7a0', fontsize='30', fontname='Helvetica-Bold')

# --- Global Connections ---
dot.edge('Shock', 'Q1_Model')
dot.edge('Shock', 'Q2_Model')
dot.edge('Shock', 'Q3_Model')
dot.edge('Shock', 'Q4_Model')
dot.edge('Shock', 'Q5_Model')

# Feed outputs to final
dot.edge('Q1_Out', 'Final')
dot.edge('Q2_Out', 'Final')
dot.edge('Q3_Out', 'Final')
dot.edge('Q4_Out', 'Final')
dot.edge('Q5_Out', 'Final')

# Render
output_path = os.path.join(OUTPUT_DIR, 'Figure_0_Framework_Maximum_Font')
dot.render(output_path, view=True, cleanup=True)
print(f"Flowchart saved to {output_path}.png")
```

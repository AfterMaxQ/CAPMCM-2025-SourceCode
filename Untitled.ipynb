```python
import pandas as pd

# 存储所有 DataFrame 的字典
dataframes = {}

# 定义年份范围 (2015 到 2024)
for year in range(2015, 2025):
    # **【关键修改】** 修正文件路径和文件名
    # 路径：当前目录下的 'Tariff Data/' 文件夹
    # 文件名：'tariff_database_' + 年份 + '.xlsx'
    file_path = f"./Tariff Data/tariff_data_{year}/tariff_database_{year}.xlsx"
    
    try:
        # 读取 Excel 文件，并将结果存储在字典中
        dataframes[year] = pd.read_excel(file_path)
        print(f"成功读取文件: {file_path}")
    except FileNotFoundError:
        print(f"错误: 找不到文件 {file_path}")
    except Exception as e:
        print(f"读取文件 {file_path} 时发生错误: {e}")

# --------------------------------------------------
# 后续操作保持不变：

# 示例：访问 DataFrame
if 2015 in dataframes:
    df_2015 = dataframes[2015]
    print(f"\n2015 数据加载成功，形状：{df_2015.shape}")

# 示例：遍历 DataFrame
print("\n# --- 遍历所有数据 ---")
for year, df in dataframes.items():
    print(f"{year} 数据前两行:")
    print(df.head(2))
```

```python
col_list_per_year = []

for year, df in dataframes.items():
    print(f"--- 年份: {year} ---")
    # 现在 df 变量是一个 DataFrame，可以安全地访问 .columns
    col_list_per_year.append({year : df.columns})
    print(df.columns)
```

```python
import pprint

# 增加最大显示列数，例如设置为 500，以确保所有列名都被显示
pd.set_option('display.max_columns', 500)
# 确保列名不会换行显示，设置宽度为 None
pd.set_option('display.width', None)

pprint.pprint(col_list_per_year)
```

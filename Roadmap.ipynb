```python
from graphviz import Digraph
import os

# Configuration
OUTPUT_DIR = r"G:\jupyter\2025APMCM\Cleaned_Data"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Create Digraph
dot = Digraph(comment='Systemic Modeling Framework', format='png')
dot.attr(rankdir='TB', size='12,15', dpi='300', compound='true', splines='ortho')

# Global Node Style
dot.attr('node', shape='box', style='filled', fontname='Times-Roman', fontsize='24', margin='0.3,0.2')
dot.attr('edge', fontname='Times-Roman', fontsize='18', color='#555555')

# --- 1. Phase I: Data Forensics (Foundation) ---
with dot.subgraph(name='cluster_0') as c:
    c.attr(label='Phase I: Data Forensics & Shock Construction (Section 5)', style='filled', color='#f2f2f2', fontsize='28', fontname='Times-Bold')
    
    c.node('Raw', 'Raw Data Ingestion\n(USITC, FRED, UN Comtrade)', fillcolor='#e1e1e1')
    c.node('Clean', 'ETL & Anomaly Detection\n(Isolation Forest Algorithm)', fillcolor='#ffffff', color='#333333')
    c.node('Baseline', 'Baseline Verification\n(Pre-Shock Tariff ≈ 3.78%)', fillcolor='#ffffff', color='#333333')
    # Use "Exogenous Input" to match paper logic
    c.node('Shock', 'Policy Shock Function S(·)\n(Constructed 10% Floor)', shape='hexagon', fillcolor='#ffddc1', color='#d35400')
    
    c.edge('Raw', 'Clean')
    c.edge('Clean', 'Baseline')
    c.edge('Baseline', 'Shock', label='Regime Shift')

# --- 2. Phase II: Sectoral Transmission (Micro/Meso) ---
with dot.subgraph(name='cluster_1') as c:
    c.attr(label='Phase II: Sectoral Transmission Mechanisms (Section 6-8)', style='dashed', color='#2980b9', fontsize='28', fontname='Times-Bold')
    
    # Q1: Agriculture
    c.node('Q1_Model', 'Q1: Soybean Ecology\n(Endogenous Lotka-Volterra)', fillcolor='#d4e6f1')
    c.node('Q1_Out', 'Output: "Stagnation Trap"\n& Trade Diversion', fillcolor='#ffffff', style='dashed')
    
    # Q2: Automotive
    c.node('Q2_Model', 'Q2: Auto Supply Chain\n(VARX & Cost Matrix)', fillcolor='#d4e6f1')
    c.node('Q2_Out', 'Output: "Tariff Jumping"\n(Optimal FDI Strategy)', fillcolor='#ffffff', style='dashed')
    
    # Q3: Semiconductors
    c.node('Q3_Model', 'Q3: Chip Sovereignty\n(Macro-Linked Game Theory)', fillcolor='#d4e6f1')
    c.node('Q3_Out', 'Output: "Security Paradox"\n(Nash Equilibrium)', fillcolor='#ffffff', style='dashed')

    # Internal Edges
    c.edge('Q1_Model', 'Q1_Out')
    c.edge('Q2_Model', 'Q2_Out')
    c.edge('Q3_Model', 'Q3_Out')

# --- 3. Phase III: Macro Synthesis (Macro) ---
with dot.subgraph(name='cluster_2') as c:
    c.attr(label='Phase III: Systemic Aggregation (Section 9-10)', style='dashed', color='#27ae60', fontsize='28', fontname='Times-Bold')
    
    # Q4: Fiscal
    c.node('Q4_Model', 'Q4: Fiscal Forecast\n(Hybrid SARIMAX-LSTM)', fillcolor='#d5f5e3')
    c.node('Q4_Out', 'Output: Dynamic Laffer Curve\n(Elasticity Trap)', fillcolor='#ffffff', style='dashed')
    
    # Q5: Macro
    c.node('Q5_Model', 'Q5: Reshoring Evaluation\n(Entropy Weight Index)', fillcolor='#d5f5e3')
    c.node('Q5_Out', 'Output: "False Reshoring"\n& Stagflation Risk', fillcolor='#ffffff', style='dashed')

    c.edge('Q4_Model', 'Q4_Out')
    c.edge('Q5_Model', 'Q5_Out')

# --- 4. Final Conclusion ---
dot.node('Final', 'Global Policy Evaluation\n(Suboptimal Nash Equilibrium)', shape='Mdiamond', fillcolor='#fad7a0', fontsize='28', fontname='Times-Bold')

# --- Global Connections (The "Systemic" Link) ---
# Connecting Shock to all models implies the "Systemic" nature
dot.edge('Shock', 'Q1_Model')
dot.edge('Shock', 'Q2_Model')
dot.edge('Shock', 'Q3_Model')
dot.edge('Shock', 'Q4_Model')
dot.edge('Shock', 'Q5_Model')

# Connecting Outputs to Final Conclusion
dot.edge('Q1_Out', 'Final')
dot.edge('Q2_Out', 'Final')
dot.edge('Q3_Out', 'Final')
dot.edge('Q4_Out', 'Final')
dot.edge('Q5_Out', 'Final')

# Render
output_path = os.path.join(OUTPUT_DIR, 'Figure_1_Systemic_Framework')
dot.render(output_path, view=True, cleanup=True)
print(f"Flowchart saved to {output_path}.png")
```
